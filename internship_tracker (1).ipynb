```python
# インターン管理システム

import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

data = [
    ["SCSK", "SIer", "SE", "WEBテスト受験済", "2026-06-18", "高", 5, "SIer志望と接客経験を活かす"],
    ["NECネッツエスアイ", "ICT・SIer", "技術職", "適性検査受験済", "2026-06-17", "高", 5, "社会インフラを支えるICTに興味"],
    ["SKY", "IT", "エンジニア", "結果待ち", "2026-06-16", "中", 4, "開発職への興味"],
    ["CTC", "SIer", "SE", "不合格", "2026-06-08", "中", 4, "秋冬インターンで再挑戦"],
    ["TIS", "SIer", "SE", "不合格", "2026-07-10", "高", 5, "金融・決済など社会基盤に関心"]
]

columns = ["企業名", "業界", "応募職種", "選考状況", "ES締切", "優先度", "志望度", "メモ"]

df = pd.DataFrame(data, columns=columns)

df
```

```python
# 企業を追加する関数

def add_company(company, industry, job, status, deadline, priority, motivation, memo):
    global df

    new_data = {
        "企業名": company,
        "業界": industry,
        "応募職種": job,
        "選考状況": status,
        "ES締切": deadline,
        "優先度": priority,
        "志望度": motivation,
        "メモ": memo
    }

    df = pd.concat([df, pd.DataFrame([new_data])], ignore_index=True)
    return df
```

```python
# 追加の例

add_company(
    "BIPROGY",
    "SIer",
    "SE",
    "未応募",
    "2026-07-20",
    "中",
    4,
    "社会を支えるITに興味"
)
```

```python
# 選考状況ごとの件数を確認

status_count = df["選考状況"].value_counts()
status_count
```

```python
# グラフで可視化

status_count.plot(kind="bar", figsize=(8, 5))

plt.title("Selection Status Count")
plt.xlabel("Status")
plt.ylabel("Number of Companies")
plt.xticks(rotation=45)
plt.show()
```

```python
# 志望度が高い順に並び替え

df.sort_values(by="志望度", ascending=False)
```

```python
# ES締切までの日数を計算

df["ES締切"] = pd.to_datetime(df["ES締切"])
today = pd.to_datetime(datetime.today().date())

df["締切までの日数"] = (df["ES締切"] - today).dt.days

df.sort_values(by="締切までの日数")
```

```python
# CSVファイルとして保存

df.to_csv("internship_tracker.csv", index=False, encoding="utf-8-sig")

print("保存完了：internship_tracker.csv")
```
